```python
def get_one_movie(movie_id):
    link = f"https://www.netflix.com/hu/title/{movie_id}"
    print(link)
    t = read_html(link)
    data = t.find_all('script', {'type': 'application/ld+json'})
    print(data)
    movie_list = json.loads(data[0].text)
    return (movie_list)
```
